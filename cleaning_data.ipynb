```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

```python
# The default port used by MongoDB is 27017
# https://docs.mongodb.com/manual/reference/default-mongodb-port/
mongo = MongoClient(port=27017)

# Declare the database
db = mongo.project3
```

```python
db.list_collection_names()
```

```python
# Declare the collections
population = db['population']
gdp = db['gdp']
location = db['location']
```

```python
pprint(population.find_one())
```

```python
pprint(gdp.find_one())
```

```python
results_pop = list(population.find())
results_gdp = list(gdp.find())
results_loc = list(location.find())
```

```python
pop_df = pd.DataFrame(results_pop)
gdp_df = pd.DataFrame(results_gdp)
loc_df = pd.DataFrame(results_loc)
```

```python
pop_df.head()
```

```python
gdp_df.head()
```

```python
gdp_df = gdp_df[['Country Name', 'Country Code', '2022 GPD']]
```

```python
pop_df = pop_df[['Country Name', 'Country Code', '2022 POP']]
```

```python
pop_df
```

```python
gdp_df = gdp_df.rename(columns = {'2022 GPD': '2022 GDP'})
```

```python
gdp_df.head()
```

```python
gdp_df = gdp_df[gdp_df['2022 GDP'].notnull()]
gdp_df.head()
```

```python
merge_df = pd.merge(pop_df, gdp_df, on='Country Code')
merge_df.head()
```

```python
merge_df = merge_df.rename(columns = {'Country Name_x': 'Country Name'}).drop(columns = ['Country Name_y'])
merge_df.head()
```

```python
loc_df.head()
```

```python
loc_df = loc_df.drop(columns=['_id', 'Alpha-2 code', 'Numeric code', 'Country'])
loc_df
```

```python
loc_df = loc_df.rename(columns = {'Alpha-3 code': 'Country Code', 'Latitude (average)': 'Latitude', 'Longitude (average)': 'Longitude'})
loc_df
```

```python
final_df = pd.merge(merge_df, loc_df, on='Country Code')
final_df.head()
```

```python
final_df
```

```python
pop_df.to_json("Cleaned_data/population.json")
gdp_df.to_json("Cleaned_data/gdp.json")
loc_df.to_json("Cleaned_data/location.json")
final_df.to_json("Cleaned_data/final.json")
```
