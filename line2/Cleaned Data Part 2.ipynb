```python
import pandas as pd
import csv
```

```python
# Open the CSV file
with open('Country GDP.csv', 'r', newline='') as csvfile:
    csvreader = csv.reader(csvfile)
    
    # Separate the first row as column headers
    column_headers = next(csvreader)
    
    # Convert the remaining CSV data to a list of rows
    rows = [row for row in csvreader]

# Convert the list of rows to a DataFrame with column headers
gdp = pd.DataFrame(rows, columns=column_headers)

# Display the DataFrame
gdp
```

```python
# Open the CSV file
with open('Country pop.csv', 'r', newline='') as csvfile:
    csvreader = csv.reader(csvfile)
    
    # Separate the first row as column headers
    column_headers = next(csvreader)
    
    # Convert the remaining CSV data to a list of rows
    rows = [row for row in csvreader]

# Convert the list of rows to a DataFrame with column headers
pop = pd.DataFrame(rows, columns=column_headers)

# Display the DataFrame
pop
```

```python
pop.rename(columns={"锘緾ountry Name": "Name"}, inplace=True)
columns_to_drop = ['Indicator Name', 'Indicator Code', 'Country Code']
pop = pop.drop(columns=columns_to_drop)
pop
```

```python
gdp.rename(columns={"锘緾ountry Name": "Name"}, inplace=True)
columns_to_drop = ['Indicator Name', 'Indicator Code', 'Country Code']

gdp = gdp.drop(columns=columns_to_drop)
gdp
```

```python
pop.to_json("pop.json", orient = 'records')
gdp.to_json("gdp.json", orient = 'records')
```

```python
pop.to_csv("pop.csv")
gdp.to_csv("gdp.csv")
```

```python
merged_df = pd.merge(pop, gdp, on="Name", how="inner")
merged_df
```
