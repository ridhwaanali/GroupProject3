```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

```python
# The default port used by MongoDB is 27017
# https://docs.mongodb.com/manual/reference/default-mongodb-port/
mongo = MongoClient(port=27017)

# Declare the database
db = mongo.project3
```

```python
db.list_collection_names()
```

```python
# Declare the collections
population = db['population']
gdp = db['gdp']
```

```python
pprint(population.find_one())
```

```python
pprint(gdp.find_one())
```

```python
results_pop = list(population.find())
results_gdp = list(gdp.find())
```

```python
pop_df = pd.DataFrame(results_pop)
gdp_df = pd.DataFrame(results_gdp)
```

```python
pop_df.head()
```

```python
gdp_df.head()
```

```python
gdp_df = gdp_df.drop(columns = ['_id', 'Indicator Name', 'Indicator Code'])
```

```python
pop_df = pop_df.drop(columns = ['_id', 'Indicator Name', 'Indicator Code'])
```

```python
pop_df
```

```python
merge_df = pd.merge(pop_df, gdp_df, on='Country Code')
merge_df.head()
```

```python
merge_df = merge_df.rename(columns = {'Country Name_x': 'Country Name'})
merge_df.head()
```

```python
pop_df.to_json("Cleaned_data/population.json")
gdp_df.to_json("Cleaned_data/gdp.json")
merge_df.to_json("Cleaned_data/merged.json")
```
